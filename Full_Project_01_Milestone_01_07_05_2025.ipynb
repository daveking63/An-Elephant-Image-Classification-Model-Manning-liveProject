```python
# ======================================================
# TODO 1: Import necessary libraries.
# Hint: You may need os, shutil, or similar libraries to handle files and folders.
# ======================================================

import os
import zipfile
import matplotlib.pyplot as plt
import random
from pathlib import Path
from PIL import Image
```

```python
# ======================================================
# TODO 2: Download the provided dataset.
# (You can download it manually using the link shared in the project instructions.)
# After downloading, extract the dataset and inspect its folder structure.
# ======================================================

# NOTE: Replace 'your_dataset.zip' with the path to your downloaded zip file
zip_path = "your_dataset.zip"  # Update this to your actual downloaded zip path
extract_path = "elephant_dataset"  # Folder where data will be extracted

# Extract the dataset
if not os.path.exists(extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)

print(f"Dataset extracted to: {extract_path}")
```

```python
# ======================================================
# TODO 3: Inspect the dataset structure.
# Check if the data is divided into training and testing folders.
# Each should have two subfolders: 'asian' and 'african'.
# ======================================================

# Define paths
train_dir = Path(extract_path) / "train"
test_dir = Path(extract_path) / "test"

# Print the contents
print("Training data folders:", os.listdir(train_dir))
print("Testing data folders:", os.listdir(test_dir))

# Further check inside training folders
print("\nSample files in 'train/asian':", os.listdir(train_dir / "Asian")[:5])
print("Sample files in 'train/african':", os.listdir(train_dir / "African")[:5])
```

```python
# ======================================================
# TODO 4: Verify a few sample images from each category.
# Display 2-3 random images from 'asian' and 'african' classes to visually verify correctness.
# ======================================================

# Function to display random images from a class
def display_random_images(folder_path, label, num_images=3):
    images = os.listdir(folder_path)
    selected_images = random.sample(images, num_images)

    plt.figure(figsize=(12, 4))
    for idx, img_name in enumerate(selected_images):
        img_path = folder_path / img_name
        img = Image.open(img_path)

        plt.subplot(1, num_images, idx + 1)
        plt.imshow(img)
        plt.title(f"{label}")
        plt.axis("off")
    plt.show()

# Display random images from Asian elephants
print("\nSample Asian Elephant Images:")
display_random_images(train_dir / "Asian", "Asian Elephant")

# Display random images from African elephants
print("\nSample African Elephant Images:")
display_random_images(train_dir / "African", "African Elephant")
```
